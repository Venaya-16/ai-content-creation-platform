```python
import google.generativeai as genai
from google.colab import userdata
import gradio as gr
```

```python
import google.generativeai as genai
genai.configure(api_key=userdata.get('GOOGLE_API_KEY'))
model = genai.GenerativeModel('gemini-3.5-flash')
```

```python
def ask_gemini(prompt):

    response = model.generate_content(
        contents=prompt
    )

    return response.text
```

```python
def generate_content(tool, topic, audience, tone):

    if tool == "Blog Generator":

        prompt = f"""
You are an expert SEO content writer.

Write a well-structured blog.

Topic: {topic}

Target Audience: {audience}

Tone: {tone}

Requirements:
- Catchy title
- Introduction
- 3-5 headings
- Conclusion
- SEO friendly
- Around 600 words
"""

    elif tool == "Caption Generator":

        prompt = f"""
Generate an engaging social media caption.

Topic: {topic}

Audience: {audience}

Tone: {tone}

Requirements:
- Platform-neutral caption
- Include emojis where appropriate
- Include 5-8 relevant hashtags
"""

    elif tool == "Ad Copy Generator":

        prompt = f"""
Generate a persuasive advertisement.

Product/Service:
{topic}

Target Audience:
{audience}

Tone:
{tone}

Requirements:
- Headline
- Description
- Strong Call-to-Action
"""

    else:

        prompt = f"""
You are an SEO expert.

Topic:
{topic}

Primary Audience:
{audience}

Requirements:
Generate:

1. SEO Title

2. Meta Description

3. 10 SEO Keywords

4. Suggested Blog Headings

5. URL Slug

Tone:
{tone}
"""

    return ask_gemini(prompt)
```

```python
print(generate_content("Blog Generator", "The Future of AI in Education", "Educators and Students", "Informative"))
```

```python
demo = gr.Interface(

    fn=generate_content,

    inputs=[

        gr.Dropdown(

            [

                "Blog Generator",

                "Caption Generator",

                "Ad Copy Generator",

                "SEO Optimizer"

            ],

            label="Choose Tool"

        ),

        gr.Textbox(

            label="Topic / Product",

            placeholder="Enter your topic..."

        ),

        gr.Textbox(

            label="Target Audience",

            placeholder="Students, Developers, Businesses..."

        ),

        gr.Dropdown(

            [

                "Professional",

                "Friendly",

                "Formal",

                "Persuasive",

                "Creative"

            ],

            label="Writing Tone"

        )

    ],

    outputs=gr.Markdown(label="Generated Content"),

    title="✍️ AI Content Creation Platform",

    description="Generate blogs, captions, ad copies and SEO content using Google Gemini."

)
```

```python
demo.launch()
```
